```python
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('city'):
        m = np.mean(subdf.pricePerSquareFoot)
        st = np.std(subdf.pricePerSquareFoot)
        reduced_df = subdf[(subdf.pricePerSquareFoot>(m-st)) & (subdf.pricePerSquareFoot<=(m+st))]
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out
new_df4 = remove_pps_outliers(new_df3)
new_df4.shape
```
