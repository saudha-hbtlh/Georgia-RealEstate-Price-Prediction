```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
import pickle
```

```python
model=pd.read_pickle('Georgia_home_prices_model.pickle')
model
```

```python
new_df4 = pd.read_csv('real_estate_new_real_file.csv')
```

```python
X_digits =new_df4.drop(['price'],axis='columns')
Y_digits =new_df4.price
```

```python
	from sklearn.preprocessing import LabelEncoder 
	le_city=LabelEncoder()
	le_county=LabelEncoder()
	le_homeType=LabelEncoder()
	le_levels=LabelEncoder()
	
	new_df4['city_en']=le_city.fit_transform(new_df4['city'])
	new_df4['county_en']=le_county.fit_transform(new_df4['county'])
	new_df4['hometype_en']=le_homeType.fit_transform(new_df4['homeType'])
	new_df4['levels_en']=le_levels.fit_transform(new_df4['levels'])
new_df4
```

```python
new_df6=new_df4.drop(['city','county','homeType','levels'],axis='columns')
```

```python
X_digits =new_df6.drop(['price'],axis='columns')
Y_digits =new_df6.price
X_digits
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_digits, Y_digits, test_size=0.2, random_state=42)
```

```python
import numpy as np
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split

# Generate some random data for demonstration


# Create a Linear Regression model


# Define the hyperparameters and their possible values
param_grid = {
    'fit_intercept': [True, False],
}

# Create the GridSearchCV object
grid_search = GridSearchCV(model, param_grid, cv=5, scoring='neg_mean_squared_error')

# Fit the model to the data and find the best parameters
grid_search.fit(X_train, y_train)

# Print the best parameters and corresponding score
print("Best Parameters: ", grid_search.best_params_)
print("Best Score: ", -grid_search.best_score_)

# Evaluate the model with the best parameters on the test set
best_model = grid_search.best_estimator_
test_score = best_model.score(X_test, y_test)
print("Test Set Score: ", test_score)
```
